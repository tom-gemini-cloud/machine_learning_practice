# Supervised Learning Classification


## Discriminant Analysis

Discriminant Analysis is a family of supervised classification methods that model each class’s distribution and then use Bayes’ rule to assign the most likely class.

### Core idea (Discriminant Analysis)

Assume each data point is a feature vector:

- **x** ∈ ℝ^d (a vector of _d_ real-valued features)

Assume the features come from a **class-conditional density**:

- **p(x | y = k)** (the probability density of seeing x if the class is k)

and each class has a **prior probability**:

- **π_k = P(y = k)** (how common class k is overall)

### Prediction rule

Predict the class **ŷ** by choosing the class _k_ that maximizes:

\[
\hat{y} = \arg\max_k \left[ \log p(x \mid y = k) + \log \pi_k \right].
\]

### Common modeling choice

Most commonly, \( p(x \mid y = k) \) is modeled as a **multivariate Gaussian**.
